```python
import numpy as np
import cv2
```

```python
img_luv = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2Luv)
```

```python
img_luv
```

```python
plt.imshow(img_luv)
plt.title('LUV image')
```

```python
img_luv.shape
```

```python
gap = 1
img_luv_new = np.asarray([[list(img_luv[i, j, :]) for j in range(0,img_luv.shape[1], gap)]
    for i in np.arange(0,img_luv.shape[0], gap)])
```

```python
img_luv_new
```

```python
plt.imshow(img_luv_new)
```

```python
l, u, v = [], [], []
for i in range(0,img_luv_new.shape[0]):
    for j in np.arange(0,img_luv_new.shape[1]):
        l.append(img_luv_new[i,j,0])
        u.append(img_luv_new[i,j,1])
        v.append(img_luv_new[i,j,2])
```
